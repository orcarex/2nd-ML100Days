```python
import os
import numpy as np
import pandas as pd
```

```python
# 設定 data_path
dir_data = './'

f_app = os.path.join(dir_data, 'train.csv')

print('Path of read in data: %s' % (f_app))

x_train = pd.read_csv(f_app,header=None)
```

```python
y_train = pd.read_csv('trainLabels.csv',header=None)
# y_train
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split

# 切分訓練集/測試集
x_train, x_test, y_train, y_test = train_test_split( x_train, y_train, test_size=0.25, random_state=4)

# 建立模型
reg = GradientBoostingClassifier()

# 訓練模型
reg.fit(x_train, y_train)

# 預測測試集
y_pred = reg.predict(x_test)
```

```python
from sklearn import metrics
```

```python
acc = metrics.accuracy_score(y_test, y_pred)
print("Acuuracy: ", acc)
```

```python
from sklearn.model_selection import KFold, GridSearchCV
# 設定要訓練的超參數組合
n_estimators = [50, 100, 150, 200]
max_depth = [1, 3, 5, 7, 9]
param_grid = dict(n_estimators=n_estimators, max_depth=max_depth)

## 建立搜尋物件，放入模型及參數組合字典 (n_jobs=-1 會使用全部 cpu 平行運算)
grid_search = GridSearchCV(reg, param_grid, scoring="neg_mean_squared_error", n_jobs=-1, verbose=1)

# 開始搜尋最佳參數
grid_result = grid_search.fit(x_train, y_train)

# 預設會跑 3-fold cross-validadtion，總共 9 種參數組合，總共要 train 27 次模型
```

```python
# 印出最佳結果與最佳參數
print("Best Accuracy: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
grid_result.best_params_
```

```python
# 使用最佳參數重新建立模型
reg_bestparam = GradientBoostingClassifier(max_depth=grid_result.best_params_['max_depth'],
                                           n_estimators=grid_result.best_params_['n_estimators'])

# 訓練模型
reg_bestparam.fit(x_train, y_train)

# 預測測試集
y_pred = reg_bestparam.predict(x_test)
```

```python
# 調整參數後約可降至 0.05 的 MSE
print(metrics.mean_squared_error(y_test, y_pred))
```

```python
acc = metrics.accuracy_score(y_test, y_pred)
print("Acuuracy: ", acc)
```

```python
need_pred = pd.read_csv('test.csv',header=None)
need_pred.head()
```

```python
predictions = reg_bestparam.predict(need_pred)
prediction = pd.DataFrame(predictions, columns=['Solution'])
prediction
prediction.to_csv('prediction.csv')
```
